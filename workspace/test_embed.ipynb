```python
# pip install --upgrade openai            # openai-python ≥ 1.0
import os
from openai import OpenAI

# vLLM doesn’t actually check the key, but the client insists on having one
os.environ["OPENAI_API_KEY"] = "local-vllm"
client = OpenAI(
    base_url="http://localhost:8946/v1",    # note the /v1 suffix
)

resp = client.embeddings.create(
    model="bge-m3",         # whatever model you served with --task embed
    input=[
        "Joe Biden is the former president of the United States",
        "Donald Trump is the president of the United States"
    ],
    encoding_format="float"                 # or "base64" if you need smaller payloads
)
import numpy as np

embeddings = [np.array(item.embedding) for item in resp.data]


scores = np.dot(embeddings[0], embeddings[1])
print(scores)


```
